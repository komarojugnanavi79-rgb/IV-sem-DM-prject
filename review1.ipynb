```python
import pandas as pd

# -------------------------------
# Function to clean one dataset
# -------------------------------
def clean_dataset(file_name, dataset_name):
    df = pd.read_csv(file_name)

    # 1. Standardize column names
    df.columns = df.columns.str.strip().str.lower().str.replace(" ", "_")

    # 2. Remove duplicate rows
    df = df.drop_duplicates()

    # 3. Handle missing values
    # Numeric  mean
    num_cols = df.select_dtypes(include=["int64", "float64"]).columns
    df[num_cols] = df[num_cols].fillna(df[num_cols].mean())

    # Categorical  mode
    cat_cols = df.select_dtypes(include=["object"]).columns
    df[cat_cols] = df[cat_cols].fillna(df[cat_cols].mode().iloc[0])

    # 4. Reset index
    df.reset_index(drop=True, inplace=True)

    # 5. Show output
    print(f"\nCleaned {dataset_name}:")
    print(df.head())
    print("\nMissing values check:")
    print(df.isnull().sum())

    return df


# -------------------------------
# Clean datasets individually
# -------------------------------
df1 = clean_dataset("/content/data set 1.csv", "Dataset 1")
df2 = clean_dataset("/content/data set 2.csv", "Dataset 2")
df3 = clean_dataset("/content/data set 3.csv", "Dataset 3")
```

```python
import pandas as pd
#cleaning
# Load datasets
df1 = pd.read_csv("/content/data set 1.csv")
df2 = pd.read_csv("/content/data set 2.csv")
df3 = pd.read_csv("/content/data set 3.csv")

# -------------------------------
# 1. Standardize column names
# -------------------------------
def clean_columns(df):
    df.columns = df.columns.str.strip().str.lower().str.replace(" ", "_")
    return df

df1 = clean_columns(df1)
df2 = clean_columns(df2)
df3 = clean_columns(df3)

# -------------------------------
# 2. Remove duplicate rows
# -------------------------------
df1 = df1.drop_duplicates()
df2 = df2.drop_duplicates()
df3 = df3.drop_duplicates()

# -------------------------------
# 3. Handle missing values
# -------------------------------
def handle_missing(df):
    # Numeric columns  mean
    num_cols = df.select_dtypes(include=["int64", "float64"]).columns
    df[num_cols] = df[num_cols].fillna(df[num_cols].mean())

    # Categorical columns  mode
    cat_cols = df.select_dtypes(include=["object"]).columns
    df[cat_cols] = df[cat_cols].fillna(df[cat_cols].mode().iloc[0])

    return df

df1 = handle_missing(df1)
df2 = handle_missing(df2)
df3 = handle_missing(df3)

# -------------------------------
# 4. Reset index
# -------------------------------
df1.reset_index(drop=True, inplace=True)
df2.reset_index(drop=True, inplace=True)
df3.reset_index(drop=True, inplace=True)

# -------------------------------
# 5. Show cleaned outputs
# -------------------------------
print("Cleaned Dataset 1:")
print(df1.head())

print("\nCleaned Dataset 2:")
print(df2.head())

print("\nCleaned Dataset 3:")
print(df3.head())

# -------------------------------
# 6. Verify no null values
# -------------------------------
print("\nMissing values check:")
print("Dataset 1:\n", df1.isnull().sum())
print("Dataset 2:\n", df2.isnull().sum())
print("Dataset 3:\n", df3.isnull().sum())
```

```python
import pandas as pd
#integration
# Load cleaned datasets
# Use the already cleaned df1, df2, df3 from the previous cell
# df1 = pd.read_csv("dataset1_cleaned.csv")
# df2 = pd.read_csv("dataset2_cleaned.csv")
# df3 = pd.read_csv("dataset3_cleaned.csv")

# Make sure student_id column is same in all datasets
df1.rename(columns={"studentid": "student_id"}, inplace=True)
df2.rename(columns={"studentid": "student_id"}, inplace=True)
df3.rename(columns={"studentid": "student_id"}, inplace=True)

# Integrate dataset 1 and 2
integrated_df = pd.merge(df1, df2, on="student_id", how="inner")

# Integrate with dataset 3
final_integrated_df = pd.merge(integrated_df, df3, on="student_id", how="inner")

# Reset index
final_integrated_df.reset_index(drop=True, inplace=True)

# Show integrated dataset
print("Integrated Dataset:")
print(final_integrated_df.head())

print("\nDataset Shape:", final_integrated_df.shape)
```

```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder, MinMaxScaler
#transformation
# Load integrated dataset
df = final_integrated_df

# -------------------------------
# 1. Separate column types
# -------------------------------
categorical_cols = df.select_dtypes(include=["object"]).columns
numerical_cols = df.select_dtypes(include=["int64", "float64"]).columns

# Remove student_id from transformation
numerical_cols = numerical_cols.drop("student_id", errors="ignore")

# -------------------------------
# 2. Convert categorical to numeric
# -------------------------------
encoder = LabelEncoder()
for col in categorical_cols:
    df[col] = encoder.fit_transform(df[col])

# -------------------------------
# 3. Normalize numerical columns
# -------------------------------
scaler = MinMaxScaler()
df[numerical_cols] = scaler.fit_transform(df[numerical_cols])

# -------------------------------
# 4. Show transformed data
# -------------------------------
print("Transformed Dataset:")
print(df.head())

print("\nDataset Shape:", df.shape)
print("\nData Types:")
print(df.dtypes)
```

```python
import pandas as pd
from sklearn.decomposition import PCA

# Load transformed dataset
df = df

# -------------------------------
# 1. Separate features and ID
# -------------------------------
student_id = df["student_id"]
features = df.drop(columns=["student_id"])

# -------------------------------
# 2. Apply PCA
# -------------------------------
# Reduce to 5 components (you can change this)
pca = PCA(n_components=5)
reduced_data = pca.fit_transform(features)

# -------------------------------
# 3. Create reduced dataframe
# -------------------------------
reduced_df = pd.DataFrame(
    reduced_data,
    columns=["PC1", "PC2", "PC3", "PC4", "PC5"]
)

reduced_df["student_id"] = student_id

# -------------------------------
# 4. Show reduced dataset
# -------------------------------
print("Reduced Dataset:")
print(reduced_df.head())

print("\nReduced Dataset Shape:", reduced_df.shape)

print("\nExplained Variance Ratio:")
print(pca.explained_variance_ratio_)
```
